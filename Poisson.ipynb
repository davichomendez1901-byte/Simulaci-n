En este trabajo exploraremos y programaremos la distribución de Poisson. La cual es una ddistribición de probabilidad discreta, la cual describe la probabilidad de que un evento ocurra un número determinado de veces en un intervalo de tiempo fijo o de espacio (en especial cuando dichos eventos son raros o bien independientes). 
Por ejemplo, podemos aplicar esta distribución en el número de llamadas a un centro telefónico, el número bacterias en un voolumen de agua, o el número de eventos en un area determinada.

$$
P(X = k) = \frac{\lambda^k e ^{-\lambda}}{k!}, \quad k = 0,1,2, \dots 
$$

Esta formula es la que utilizaremos para calcular la probabilidad de que ocurra un evento en un intervaloo dado. donde lambda será nuestra tasa promedio de eventos y k es el número de eventos observados.



```python
import numpy as np
import matplotlib.pyplot as plt
import random as rm
import math
#Primero creamos la funcion Poisson
def poisson(lam, k):
    return (lam**k * np.exp(-lam)) / math.factorial(k)
#Poisson inversa:
def poisson_inversa(lam):
    u = rm.random() #Uniforme entre 0 y 1
    F = 0           # Acumulador de la funcion de distribucion
    k = 0           # Contador de eventos
    while F < u:
        F += poisson(lam, k) # Acumulamos la probabilidad de k eventos
        if u <= F:
            return k          # Devolvemos el numero de eventos
        k += 1              # Incrementamos el contador de eventos
```

```python
muestras = []
for i in range (1000):
    X = poisson_inversa(5) # Generamos 1000 muestras de la distribucion de Poisson con lambda = 5
    muestras.append(X)
```

```python
plt.hist(muestras, bins=range(0, max(muestras)+1), density=True, alpha=0.6, color='g')  
plt.title('Histograma de muestras de Poisson (lambda=5)')
plt.xlabel('Número de eventos') 
plt.ylabel('Frecuencia relativa')
plt.grid()
plt.show()
```
