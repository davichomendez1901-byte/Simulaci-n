```python
#calculamos la media muestral:
import math
n = len(datos)
#calculo de la media.
sum = 0
for u in datos:
  sum += u
  media = sum / n
print("La media muestral es:",media, ":)")
print("El tamaño de nuestra muestra es:", n,)
#Calculemos la varianza:
var = 1/12
#Definimos Z
Z = (media - 0.5)  * math.sqrt(n) / math.sqrt(var)
print("El valor de Z es:", Z)
#Dado que estamos trabajando con un alpha = 0.05, nuestro Z crítico es 1.96
Z_c=1.96
#Ahora simplemente decimos que:
if abs(Z) < Z_c:
  print("No se debe de rechazar H_0 :)")
else:
  print("Se rechaza H_0 :(")
```
