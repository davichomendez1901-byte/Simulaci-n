Esta prueba es una prueba estadística no paramnétrica que nos permite evaluar la independencia y la aleatoridad de una secuencia de datos.   
Para esta prueba se consideran  las hipotesis:
* $H_0$ : los números de la muestra ${X_i}$ son independientes.
* $H_1$ los números de la muestra ${X_i}$ no son imndependintes.  
Para ello el primer paso a realizar será construir la secuencia de signos $S$

Para $i= 2,....,n$


$$  
Si =
\begin{cases}
+, & X_i >= X_{i-1} \\
- ,& X_i =< X_{i-1} \\
\end{cases}
$$

Para posteriormente contar las rachas, las cuales son un número consecutivo de signos iguales, ejemplo:  
$+++ --- + ---$  
Hay 4 Rachas.
Una vez contado el número de rachas $C_o$ se compara con su valor esperado:

$\mu_{C_0}=\frac{2n-1}{3}$, $\sigma^2_{C_0}=\frac{16n-29}{90}$

Con un estadístico de prueba:
$Z_0=\frac{C_0-\mu_{C_0}}{\sigma_{C_0}}$ si $|Z_0| < 1.96$

```python
#colocamos nuestro conjunto de datos pseudoaleatorios y los hacemos una matriz.
import math
import numpy as np
matriz = np.array([
[0.78961, 0.05230, 0.10699, 0.55877, 0.14151],
[0.76086, 0.12079, 0.27738, 0.65726, 0.79269],
[0.80548, 0.82654, 0.29453, 0.20852, 0.42989],
[0.58518, 0.98611, 0.34488, 0.34358, 0.11537],
[0.89898, 0.57880, 0.67621, 0.05010, 0.00121],
[0.28269, 0.73059, 0.70119, 0.18284, 0.49962],
[0.38618, 0.76910, 0.68334, 0.55170, 0.10850],
[0.79882, 0.45679, 0.21631, 0.87616, 0.55743],
[0.58962, 0.33216, 0.03185, 0.61168, 0.09264],
[0.69623, 0.17028, 0.05475, 0.91512, 0.76262],
[0.29931, 0.30861, 0.83358, 0.51781, 0.03272],
[0.57410, 0.26593, 0.85903, 0.43308, 0.35286],
[0.24000, 0.65559, 0.38507, 0.90829, 0.94187],
[0.93655, 0.88809, 0.81772, 0.36982, 0.19904],
[0.54325, 0.62400, 0.09133, 0.41678, 0.33954],
[0.58244, 0.85853, 0.86752, 0.38729, 0.15506],
[0.23949, 0.53559, 0.33381, 0.49883, 0.75103],
[0.19962, 0.65002, 0.74579, 0.79113, 0.63453],
[0.19147, 0.40644, 0.08128, 0.73435, 0.22724],
[0.22287, 0.07281, 0.64183, 0.44267, 0.72102]
])
```

```python
import numpy as np
import math
#"Aplanamos la matriz por columnas."
x = matriz.flatten(order = 'F')
n = len(x)
#Generaamos nuestra secuencia de signos S(+ si sube o se mantiene y - si baja)
signos=[]
for i in range(1,n):
  if x[i] >= x[i-1]:
    signos.append("+")
  else:
    signos.append("-")
#Ahora contamos nuestro numero de rachas C_0:
C_0 = 1
for i in range(1, len(signos)):
  if signos[i] != signos[i-1]:
    C_0 += 1
#Definimos la media, la varianza y Z
mu_C0 = ((2*n)-1)/3
sig_C0cua = (((16*n)-29)/90)
sig_C0 = math.sqrt(sig_C0cua)
Z_0 = (C_0-mu_C0)/sig_C0
#con \alpha = 0.05 tenemos que:
Z_crit = 1.96
print(f"n = {n}")
print(f"C0 (rachas observadas) = {C_0}")
print(f"mu_C0 = {mu_C0:.4f}")
print(f"sigma_C0 = {sig_C0:.4f}")
print(f"Z0 = {Z_0:.4f}")

if abs(Z_0) <= Z_crit:
    print("Decisión: No se rechaza H0 (independencia).")
else:
    print("Decisión: Se rechaza H0 (no independencia).")
```
