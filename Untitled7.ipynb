```python
import numpy as np
import pandas as pd
#Convertimos la lista en un arreglo
Data = np.array(Data)
n = len(Data)
#Ordenamos los datos:
Data_ord = np.sort(Data)
#Para U(0,1),F(x)=x
F_teo = Data_ord
#Ahora construimos 1/n e (i-1)/n
i = np.arange(1,n+1)
i_abvn= i/n
i_abvn_les1 = (i-1)/n
#Hacemos el calculo de D+ y D-:
D_mas_vec = i_abvn - F_teo
D_men_vec = F_teo - i_abvn_les1
#-------------------------------------------#
D_mas = np.max(D_mas_vec)
D_men = np.max(D_men_vec)

D =  max(D_mas,D_men)
#Ahora mostramos los resultados obtenidos ejecutando:
print("D+=",D_mas)
print("D-=",D_men)
print("D=",D)
#
```
