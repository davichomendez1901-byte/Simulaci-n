En este codigo implementaremos la prueba de las frecuencias a una serie de datos dada, esto para probar si proviene de una distribución uniforme $U(0,1).$
Esta prueba comparara las frecuencias observadas en intervalos de rango $[0,1]$ con las frecuencias esperadas bajo uniformidad, utilizando el estadístico (Chi-cuadrada) para así poder medir las discrepancias entre ambas dsitribuciones.


```python
#Comenzamos coloando las datos que estaremos utilizando, para ello los definiremos como:
data = [
0.78961, 0.05230, 0.10699, 0.55877, 0.14151,
0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
0.80548, 0.82654, 0.29453, 0.20852, 0.42989,
0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
0.89898, 0.57880, 0.67621, 0.05010, 0.00121,
0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
0.38618, 0.76910, 0.68334, 0.55170, 0.10850,
0.79882, 0.45679, 0.21631, 0.87616, 0.55743,
0.58962, 0.33216, 0.03185, 0.61168, 0.09264,
0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
0.29931, 0.30861, 0.83358, 0.51781, 0.03272,
0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
0.24000, 0.65559, 0.38507, 0.90829, 0.94187,
0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
0.54325, 0.62400, 0.09133, 0.41678, 0.33954,
0.58244, 0.85853, 0.86752, 0.38729, 0.15506,
0.23949, 0.53559, 0.33381, 0.49883, 0.75103,
0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
0.19147, 0.40644, 0.08128, 0.73435, 0.22724,
0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]
#N=len(data), son 100 datos
#Ahora definimos los interavalos para poder generar la tabal de frecuecnias, la cual crearemos mediante pandas.
import pandas as pd
import numpy as np
inter = [0, 0.2, 0.4, 0.6, 0.8, 1]
#Ahora Calculamos las frecuencias observadas:
fobs, _ = np.histogram(data, inter)
#Ahora las frecuencias esperadas tomando en cuenta que tenemos 100 datos:
n=len(data)
K=5
fexp = n/K
#Contruimos la taba mediante el uso de pandas:
tabla = pd.DataFrame({
    "Intervalo": ["[0-0.2]","[0.2-0.4]","[0.4-0.6]","[0.6-0.8]","[0.8-1]"],
    "fobs": fobs,
    "fexp": [fexp]*K
})
tabla["fe"]=fexp
tabla["(fobs-fexp)^2"]=(tabla["fobs"] - fexp)**2 / fexp
tabla
```

```python
#Una vez construida la tabla de frecuencias, ahora calculamos chi cuadrada, para medir la discrepacia entre frecuencias observadas(fobs)
#y las frecuencias esperadas(fe)
chi_cuadrada = tabla['(fobs-fexp)^2'].sum()
print("el valor de chi cuadrada es:", chi_cuadrada)
#Finalmenete calculamos el calor crítico y la decison que tomaremos, para esto:
import scipy.stats as sp
alpha = .05 #Este será nuestro nivel de significancia.
g_l = K-1 #el grado de libertad según nuestros intervalos.
chi_critico = sp.chi2.ppf(1-alpha, g_l)
print("Nuestro valor critico es:", chi_critico)
#Regla de decisión:
#Si chi_cuadrada < chi_critico no se rechaza H0.
#Si chi_cuadrada > chi_critico se rechza H0. ent =>
if chi_cuadrada < chi_critico:
    print("No se rechaza H0 :)")
else:
    print("Se rechaza H0 :(")
```
